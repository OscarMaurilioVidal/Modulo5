```python
#Importamo las librerías a utilizar
import pandas as pd
```

# Normalización

```python
#Cargamos los datos del archivo separado por comas
cities_df = pd.read_csv('Cities.csv')
countries_df = pd.read_csv('countries.csv')
```

```python
temp_max = cities_df["temperature"].max()
temp_min = cities_df["temperature"].min()
(cities_df["temperature"] - temp_min)/(temp_max - temp_min)
```

#### Se añade la columna al dataframe

```python
cities_df["temperature_norm"] = (cities_df["temperature"] - temp_min)/(temp_max - temp_min)
cities_df
```

# Estandarización

```python
temp_mean = cities_df["temperature"].mean()
temp_mean
```

```python
temp_std = cities_df["temperature"].std()
temp_std
```

#### Se añade la columna al dataframe

```python
cities_df["temperature_std"] = (cities_df["temperature"] - temp_mean)/temp_std
cities_df
```

# Binning

```python
def temperature_to_warmth(temp):
    if temp<10:
        return "Frío"
    if temp<18:
        return "Templado"
    return "Cálido"

    
```

```python
cities_df["temperature"].apply(temperature_to_warmth)
```

```python
cities_df["temperature_type"] = cities_df["temperature"].apply(temperature_to_warmth)
```

```python
cities_df
```

```python
pd.get_dummies(cities_df["temperature_type"])
```
