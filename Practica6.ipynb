```python
#Importamo las librerías a utilizar
import pandas as pd
```

```python

players_df = pd.read_csv('players.csv')
teams_df = pd.read_csv('teams.csv')

union_df = teams_df.merge(players_df)
union_df
```

```python
union_df.sort_values(by = ["wins"],ascending = [False])
```

```python
union_df.sort_values(by = ["wins"],ascending = [False]).max()
```

```python
union_df.max()
```

```python
union_df.min()
```

```python
union_df.groupby(["wins","team"]).sum().sort_values(["wins"],ascending=[False])[0:3]
```

```python
union_df.sort_values(by=["minutes"],ascending=[False])[0:3]

```
